```python
x_train, x_val, y_train, y_val = train_test_split(df_train_x, df_train_y , test_size=.2)
#x_test, x_val, y_test, y_val = train_test_split(x_test, y_test, test_size=.5)
```

```python
drop_rate = 0.2
model = Sequential()
model.add(Flatten())          # input layer
model.add(Dense(units = 128, activation = 'relu'))      # hidden layer
model.add(Dropout(drop_rate))    # Overfitting 방지
model.add(Dense(units = 64, activation = 'relu'))      # hidden layer2
# model.add(Dropout(drop_rate))    # Overfitting 방지
model.add(Dense(units = 32, activation = 'relu'))      # hidden layer3
# model.add(Dropout(drop_rate))    # Overfitting 방지
model.add(Dense(units = 1, activation = 'sigmoid'))      # output layer
```

```python
model.compile(Adam(learning_rate = 0.01), "mse", 
              metrics = ["accuracy"])
```

```python
model.fit(x_train, y_train, verbose=1, epochs=10)
```

```python
fin_test = pd.read_csv("smart_farm_test.csv").drop(['Unnamed: 0', 'smart_farm.heat_supply',
                                                   'smart_farm.ventilation_temperature_control',
                                                    'smart_farm.heating_temperature_set_up'], axis=1)

# 전체열 변경 안먹혀서 하나씩 변경
colnames = [i[11:] for i in fin_test.columns]

for i in range(len(colnames)):
    fin_test.rename(columns = {fin_test.columns[i] : colnames[i]},inplace = True)
fin_test
```

```python
fin_test
```

```python
# 날찌 데이터 문자열로 바꾼 후 time 열 생성
fin_test.iloc[:, :5] = fin_test.iloc[:, :5].astype(str)
fin_test['time'] = fin_test['yy']+'-'+fin_test['mm']+'-'+fin_test['dd']+'-'+fin_test['hh']+'-'+fin_test['nn']
```

```python
# time 열 datetime 타입으로 변경
fin_test.time = pd.to_datetime(fin_test['time'],format='%Y-%m-%d-%H-%M')
fin_test = fin_test.set_index('time')

# 시간대별(분) 평균값 출력
fin_test = fin_test.resample(rule='T').mean()
fin_test
```

```python
rf = pd.read_csv('rainfall_test.csv', encoding='cp949' )
rf = rf[['일시','강수량(mm)']]
rf.columns=['time', 'rainfall']
rf['time'] = pd.to_datetime(rf['time'])
rf = rf.set_index('time')

#시간대별(분) 평균값 출력
rf = rf.resample(rule='T').mean()
rf.isnull().sum()
```

```python
fin_c = pd.concat([fin_test, rf], axis=1)
fin_c
```

```python
def fillna(df):
    # 연속형 변수 결측치 처리
    df['insolation'].fillna(df['insolation'].mean(), inplace=True)
    df['out_tmperature'].fillna(df['out_tmperature'].mean(), inplace=True)
    df['out_humidity'].fillna(df['out_humidity'].mean(), inplace=True)
    df['wind_speed'].fillna(df['wind_speed'].mean(), inplace=True)
    df['wind_direction'].fillna(df['wind_direction'].mean(), inplace=True)
    
    df['in_tmperature'].fillna(df['in_tmperature'].mean(), inplace=True)
    df['in_humidity'].fillna(df['in_humidity'].mean(), inplace=True)
    df['shield_light_h'].fillna(df['shield_light_h'].mean(), inplace=True)
    df['shield_tmperature_h'].fillna(df['shield_tmperature_h'].mean(), inplace=True)
    df['shield_energy_h'].fillna(df['shield_energy_h'].mean(), inplace=True)
    df['shield_energy_v'].fillna(df['shield_energy_v'].mean(), inplace=True)
    df['ceiling'].fillna(df['ceiling'].mean(), inplace=True)
    df['rainfall'].fillna(df['rainfall'].mean(), inplace = True)
    # 범주형 변수 결측치 처리
    df['exhaust_fan'].fillna(0, inplace=True)
    df['floating_fan'].fillna(0, inplace=True)
    df['fan_coil_b_site'].fillna(0, inplace=True)
    return df
fillna(fin_c)
fin_c.isnull().sum()
```

```python
min_max_scaler = MinMaxScaler()
min_max_scaler_fin = min_max_scaler.fit_transform(fin_c)

# array를 dataframe으로 변경
# 전체열 정규화
fin_test_normal = pd.DataFrame(min_max_scaler_fin, columns = fin_c.columns)
```
